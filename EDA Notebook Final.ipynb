# 1. Introduction

**Business Problem:**

Home Credit faces challenges in accurately predicting loan repayment ability for individuals with limited or no traditional credit history. Many underserved individuals are either rejected or subject to predatory lenders. Home Credit aims to improve its predictive models by leveraging alternative data such as telco and transactional information. This project focuses on enhancing the company's loan approval process while minimizing defaults. By improving accuracy, Home Credit can expand its client base responsibly, providing safer loan options and increasing profitability.

**Notebook Purpose:** 

The purpose of this notebook is to gain an understanding of the dataset and prepare the data set for a predictive model. This will include exploring all data availiable for the project, handling and describing missing data, exploratory analysis of the data, and a summary of findings.

5. **POS CASH Balance File:**
This dataset records monthly snapshots of point-of-sale (POS) or cash loans, including balance, payment delays, and contract status. Analyzing this data provides insights into the client's purchasing behavior and their ability to manage installment-based loans.

6. **Installments Payments File:**
Contains payment information related to previous loans, including scheduled payments, actual payments, and payment delays. This helps evaluate how punctual clients are with their repayments, providing essential data points for predicting the risk of future loan defaults.
